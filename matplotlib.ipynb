# Importacion y primeros pasos 

```python
import matplotlib.pyplot as plt
import numpy as np
```

## Graficos basicos

```python
#grafico de lineas
#datos 
x = np.linspace(0, 10, 100)
y = np.sin(x)

#crear el grafoco de lineas
plt.plot(x, y, label='Sine Wave')
plt.title('Grafico de lineas')
plt.xlabel('X')
plt.ylabel('Y')
plt.legend()
plt.show()
```

```python
#grafico de barras
#datos 
categories = ['A', 'B', 'C', 'D']
values = [10, 20, 25, 25]

#crear el grafico de barras
plt.bar(categories, values)
plt.title('Grafico de barras')
plt.xlabel('Categorias')
plt.ylabel('Valores')
plt.show()
```

```python
#grafico de dispercion
#datos
x = np.random.rand(100)
y = np.random.rand(100)

#crear el grafico de dispersion
plt.scatter(x, y)
plt.title('Grafico de dispersion')
plt.xlabel('X')
plt.ylabel('Y')
plt.show()
```

## Graficos avanzados 

```python
#grafico de pastel
#datos
labels = ['A', 'B', 'C', 'D']
sizes = [10, 20, 30, 40]

#crear el grafico de pastel
plt.pie(sizes, labels=labels, autopct='%1.1f%%')
plt.title('Grafico de pastel')
plt.show()
```

```python
#grafico de caja (Box Plot)
#datos
data = [np.random.randn(50), np.random.randn(50), np.random.randn(50)]

#crear el grafico de caja
plt.boxplot(data, label=['Grupo 1', 'Grupo 2', 'Grupo 3'])
plt.title('Grafico de caja')
plt.show()
```

```python
#grafico de superficie 3D
from mpl_toolkits.mplot3d import Axes3D

#datos
x = np.linspace(-5, 5, 100)
y = np.linspace(-5, 5, 100)
x, y = np.meshgrid(x, y)
z = np.sin(np.sqrt(x**2 + y**2))

#crear el grafico de superficie 3D
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.plot_surface(x, y, z, cmap='viridis')

ax.set_title('Grafico de superficie 3d')
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')
plt.show()

```

## Personalizacion de graficos

```python
#personalizar lineas y marcadores
#datos
x=np.linspace(0, 10, 100)
y=np.sin(x)

#crear el grafico con anotaciones y formas
plt.plot(x, y, label='Sine Wave')

#añadir anotaciones
plt.annotate('Punto Especial', xy=(5, 0), xytext=(7, 0.5),
             arrowprops=dict(facecolor='black', shrink=0.05))

#añadir forma
plt.axvspan(2, 4, color='lightgrey', alpha=0.5)

plt.title('Grafico con anotaciones y formas')
plt.xlabel('X')
plt.ylabel('Y')
plt.legend()
plt.show()
```

## Subgraficos

```python
#crear subgraficos
#datos
x = np.linspace(0, 10, 100)
y1 = np.sin(x)
y2 = np.cos(x)

#crear subgraficos
fig, axs = plt.subplots(2)

#primer subgrafico
axs[0].plot(x, y1, label='Sine Wave')
axs[0].set_title('Sine Wave')
axs[0].legend()

#segundo subgrafico
axs[1].plot(x, y2, label='Cosine Wave', color='orange')
axs[1].set_title('Cosine Wave')
axs[1].legend()

plt.tight_layout()
plt.show()
```

## Estilos y temas

```python
#usar diferentes estilos
#estilo 'ggplot'
plt.style.use('ggplot')

#datos
x = np.linspace(0, 10, 100)
y = np.sin(x)

#crear el grafico con estilo 'ggplot'
plt.plot(x, y, label='Sine Wave')
plt.title('Grafico con estilo ggplot')
plt.xlabel('X')
plt.ylabel('Y')
plt.legend()
plt.show()
```

```python
#crear un tema personalizado
custom_style = {
    'axes.edgecolor': 'black',
    'axes.facecolor': 'white',
    'grid.color': 'grey',
    'grid.linestyle': '--',
    'lines.linewidth': 2,
    'lines.color': 'blue'
}

plt.rcParams.update(custom_style)

#datos
x = np.linspace(0, 10, 100)
y = np.sin(x)

#crear el grafico 
plt.plot(x, y, label='Sine Wave')
plt.title('Grafico con tema personalizado')
plt.xlabel('X')
plt.ylabel('Y')
plt.legend()
plt.show()
```
