# Importacion y primeros pasos

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

#cargar conjunto de datos de ejemplo
df = sns.load_dataset('iris')
```

```python
#graficos de dispercion 
sns.scatterplot(data=df, x='sepal_length', y='sepal_width', hue='species')
plt.title('Grafico de dispercion')
plt.show()
```

```python
#grafico de lineas
sns.lineplot(data=df, x='sepal_length', y='sepal_width', hue='species')
plt.title('Grafico de lineas')
plt.show()
```

```python
#grafico de barras
sns.barplot(data=df, x='species', y='sepal_length')
plt.title('Gráfico de Barras')
plt.show()
```

```python
#grafico de distribucion
#histograma
sns.histplot(data=df, x='sepal_length', bins=10, kde=True)
plt.title('Histograma')
plt.show()
```

```python
#grafico de caja
sns.boxplot(data=df, x='species', y='sepal_length')
plt.title('Grafico de caja')
plt.show()
```

```python
#grafico de violin
sns.violinplot(data=df, x='species', y='sepal_length')
plt.title('Grafico de violin')
plt.show()
```

```python
#graficos relacionales
#grafico de pares
sns.pairplot(df, hue='species')
plt.show()
```

```python
from sklearn.preprocessing import LabelEncoder

# Convertir columnas categóricas a numéricas
df['species'] = LabelEncoder().fit_transform(df['species'])

#grafico de correlacion
corr = df.corr()

# Crear un heatmap de la matriz de correlación
sns.heatmap(corr, annot=True, cmap='coolwarm')
plt.title('Heatmap de Correlación')
plt.show()
```

## Personalizacion de graficos

```python
#estilos y temas
#aplicar un estilo
sns.set_style('whitegrid')

#crear un grafico con estilo aplicado
sns.scatterplot(data=df, x='sepal_length', y='sepal_width', hue='species')
plt.title('Grafico de dispercion con estilo')
plt.show()
```

```python
#ajustar el tamaño del grafico
plt.figure(figsize=(10,6))

#aplicar una palera de colores
sns.scatterplot(data=df, x='sepal_length', y='sepal_width', hue='species', palette='viridis')
plt.title('Grafico de dispercion con paleta de colores')
plt.show()
```

## FacetGrid y subgraficos

```python
# crear un FacetGrid
g  = sns.FacetGrid(df, col='species')
g.map(sns.histplot, 'sepal_length')
plt.show()
```

```python
#crear graficos con subgraficos
#crear subgrafico
fig, axs = plt.subplots(2, 2, figsize=(12, 8))

#primer subgrafico
sns.scatterplot(data=df, x='sepal_length', y='sepal_width', hue='species', ax=axs[0, 0])
axs[0, 0].set_title('Disperson')

#segundo subgrafico
sns.boxplot(data=df, x='species', y='sepal_length', ax=axs[0, 1])
axs[0, 1].set_title('Caja')

#tercer subgrafico
sns.violinplot(data=df, x='species', y='sepal_length', ax=axs[1, 0])
axs[1, 0].set_title('Violin')

#cuarto subgrafico
sns.barplot(data=df, x='species', y='sepal_length', ax=axs[1, 1])
axs[1, 1].set_title('Barras')

plt.tight_layout()
plt.show()
```
